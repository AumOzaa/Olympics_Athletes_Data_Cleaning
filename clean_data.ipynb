```python
import pandas as pd
import numpy as np
```

```python
bios = pd.read_csv('bios2.csv')
```

```python
df = bios
```

```python
df.drop('Name order',axis = 1,inplace =True)
```

```python
df.drop('Original name',axis = 1, inplace = True)
```

```python
df.drop('Other names', axis = 1, inplace =True)
df.drop('Title(s)',axis = 1, inplace = True)
```

```python
# Extract Height (ending with 'cm') and Weight (ending with 'kg') from 'Measurements'
df['Height'] = df['Measurements'].str.extract(r'(\d+)\s*cm')
df['Weight'] = df['Measurements'].str.extract(r'(\d+)\s*kg')
```

```python
df['Height'] = df['Height'].str.replace(' kg', '', regex=True)
df['Weight'] = df['Weight'].str.replace(' cm', '', regex=True)
```

```python
df['Height'] = df['Height'].astype(float)
df['Weight'] = df['Weight'].astype(float)
```

```python
#Now we will be parse out date of birh and date of death from the 'Birth' and 'Death' columns
df['Date_of_birth'] = df['Born'].str.extract(r'(\b\d{1,2} (?:January|February|March|April|May|June|July|August|September|October|November|December) \d{4}\b)')
#Using GPT to write the regex expression. Useful, and then try it out on regex101.com
```

```python
# convet karege abb hum date of birth ko datetime format mein
df['Date_of_birth'] = pd.to_datetime(df['Date_of_birth'])
```

```python
# abb zara dob mein se month vaera ka column banadete hain
df['Month_of_birth'] = df['Date_of_birth'].dt.month
```

```python
# abb hum marne ki date nikalege :
df['Date_of_death'] = df['Died'].str.extract(r'(\b\d{1,2} (?:January|February|March|April|May|June|July|August|September|October|November|December) \d{4}\b)')
```

```python
df['Date_of_death'] = pd.to_datetime(df['Date_of_death'])
```

```python
df['Moth_of_death'] = df['Date_of_death'].dt.month
```

```python
"""
And this how we clean the data 

"""

columns_to_keep = ['athlete_id','Full name','Born','NOC','Height','Weight','Date_of_birth',
                       'Month_of_birth','Date_of_death','Moth_of_death']

df = df[columns_to_keep]
```
